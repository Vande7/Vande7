````python
class NumericTransformers:
    """This class is used to apply 'Sigmoid' and 'Cube root'
    transformations to the input data (Vector)."""
    def _init_(self,*,df_hotel:pd.DataFrame) -> None:
        self.df_hotel 

    def properties(self)-> None:

        x = self.df_hotel

        print(f"shape: {x.shape}" f"\nSize: {x.size}")

    def Sigmoid_Transformer(self)-> int:
        """Returns the input value as a transformed data in sigmoid shape"""
        x = self.df_hotel
        numerator = 1
        denominator = 1 + np.exp(-x)
        x_sigmoid = numerator/denominator
        return x_sigmoid
    def Cube_Transformers(self)-> int :
        """Returns the input value as a transformed data in cube root shape"""
        x = self.df_hotel
        x_cube_root = np.cbrt(x)
        
        return x_cube_root
````

````python
# 3B Solution
# Generate random data
rng = np.random.default_rng(seed=15)
A = rng.random(size=(10, 5))

# Normalize the array
A_normalized = (A - A.min(axis=0)) / (A.max(axis=0) - A.min(axis=0))

# Print descriptive statistics of the normalized array
print("Descriptive statistics of normalized array:\n")
print("Mean:", np.mean(A_normalized))
print("Standard deviation:", np.std(A_normalized))
print("Minimum value:", np.min(A_normalized))
print("Maximum value:", np.max(A_normalized))
````
